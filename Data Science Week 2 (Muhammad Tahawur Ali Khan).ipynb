# Bag of Words Implementation with Visualization 

# 1. Importing Necessary Libraries 

```python
import string
import nltk
nltk.download('punkt')

from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS
from nltk.tokenize import word_tokenize
```

# 2. Creating Dataset

```python

data = [
    ("I love data science.", "Positive"),
    ("Machine learning is amazing.", "Positive"),
    ("I dislike bad data quality.", "Negative"),
    ("Data cleaning is boring.", "Negative"),
    ("The weather is great today.", "Positive"),
    ("I hate waiting in long lines.", "Negative"),
    ("Learning new algorithms is fun.", "Positive"),
    ("The project failed miserably.", "Negative"),
    ("I feel excited about the new features.", "Positive"),
    ("I can't stand bad customer service.", "Negative")
]
```

# 3. Pre-Processing of Data

```python

# Preprocessing function
def preprocess_text(text):
    # Step 1: Convert to lowercase
    text = text.lower()  # Makes text case-insensitive
    
    # Step 2: Remove punctuation
    text = ''.join([char for char in text if char not in string.punctuation]) 
    
    # Step 3: Tokenize the text into words
    tokens = word_tokenize(text)  # Split text into words
    
    # Step 4: Remove stop words
    tokens = [word for word in tokens if word not in ENGLISH_STOP_WORDS]  
    
    # Step 5: Join the tokens back into a single string
    return ' '.join(tokens)  # Return processed text

# Step 6: Apply preprocessing function to all sentences in the dataset
processed_data = [(preprocess_text(text), label) for text, label in data]


print(processed_data)
```

# 4. Count Vectorizer (For Bag of Words Model)

# 4.1 Importing Necessary Libraries

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import CountVectorizer
```

# 4.2 Extracting Data From Preprocessed Data & Creating Bow Feature 

```python
# Extract text data from the preprocessed dataset
texts = [text for text, label in processed_data]

# Create the BoW feature matrix
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(texts).toarray()
```

# 4.3 Visualization of Bag of Words (BoW) Feature

```python
terms = vectorizer.get_feature_names_out()
plt.figure(figsize=(10, 7))
sns.heatmap(X, cmap='YlGnBu', annot=True, fmt="d", xticklabels=terms, yticklabels=[f"Sentence {i+1}" for i in range(len(texts))])
plt.title('Term Frequency Heatmap')
plt.xlabel('Terms')
plt.ylabel('Sentences')
plt.show()
```

# 5. Word Frequency Visualization

# 5.1 Importing Library

```python
from collections import Counter
```

# 5.2 Flattening & Top 10 Common Words

```python
# Flatten the list of tokens and count frequencies
all_tokens = [word for text, label in processed_data for word in text.split()]
word_counts = Counter(all_tokens)

# Get the top 10 most common words
top_10_words = word_counts.most_common(10)
```

# 5.3 Visualization of Top 10 Common Words

```python
words, counts = zip(*top_10_words)
plt.figure(figsize=(10, 6))
plt.bar(words, counts)
plt.title('Top 10 Most Frequent Words')
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```

# 6. Classification Model 

# 6.1 Importing Necessary Libraries

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
```

# 6.2 Splitting Data Into Training and Testing Datasets

```python
# Labels (target values)
labels = [label for text, label in processed_data]

# Splitting the data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, labels, test_size=0.2, random_state=42)
```

# 6.3 Model Fitting 

```python
model = LogisticRegression()
model.fit(X_train, y_train)
```

# 6.4 Model Predictions 

```python
y_pred = model.predict(X_test)

print(y_pred)
```

# 6.5 Visualization 

```python
cm = confusion_matrix(y_test, y_pred, labels=["Positive", "Negative"])
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["Positive", "Negative"])
disp.plot(cmap='Blues')
plt.title('Confusion Matrix')
plt.show()
```
